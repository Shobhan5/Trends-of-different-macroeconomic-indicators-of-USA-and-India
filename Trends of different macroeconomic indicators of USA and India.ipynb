```python
import pandas as pd
import matplotlib.pyplot as plt
import os
# Read the CSV file
data = pd.read_csv(r"C:\Users\Shobhan Sarkar\OneDrive\Desktop\macro_assignment\unemployment_rate.csv")  # Replace "your_file_path.csv" with the actual file path

# Convert the 'DATE' column to datetime format
data['DATE'] = pd.to_datetime(data['DATE'])

# Plot the unemployment trend
plt.figure(figsize=(9, 5))
plt.plot(data['DATE'], data['UNRATE'], color='red', marker='o', linestyle='-')

# Annotate the graph with features
plt.title('USA Unemployment Trend')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)

# Add annotations for lowest and highest unemployment rates
lowest_unemployment = data.loc[data['UNRATE'].idxmin()]
highest_unemployment = data.loc[data['UNRATE'].idxmax()]

plt.annotate(f"Lowest: {lowest_unemployment['UNRATE']}% ({lowest_unemployment['DATE'].strftime('%Y-%m-%d')})",
             xy=(lowest_unemployment['DATE'], lowest_unemployment['UNRATE']),
             xytext=(lowest_unemployment['DATE'], lowest_unemployment['UNRATE'] + 1),
             arrowprops=dict(facecolor='green', shrink=0.05))

plt.annotate(f"Highest: {highest_unemployment['UNRATE']}% ({highest_unemployment['DATE'].strftime('%Y-%m-%d')})",
             xy=(highest_unemployment['DATE'], highest_unemployment['UNRATE']),
             xytext=(highest_unemployment['DATE'], highest_unemployment['UNRATE'] - 1),
             arrowprops=dict(facecolor='red', shrink=0.05))

# Show plot
#desktop_path = os.path.join(os.path.join(os.environ['USERPROFILE']), 'Desktop')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig(os.path.join(r"C:\Users\Shobhan Sarkar\OneDrive\Desktop\macro_assignment", 'unemployment_trend.png'))
print(data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

# Read the CSV file
data = pd.read_csv(r"C:\Users\Shobhan Sarkar\OneDrive\Desktop\macro_assignment\real_personal_cons_exp_per_capita.csv")

# Extract relevant columns
data['DATE'] = pd.to_datetime(data['DATE']).dt.year

# Create a wider figure
plt.figure(figsize=(10, 6))

# Plot the data
plt.plot(data['DATE'], data['personal_consumption'], marker='o', linestyle='', color='purple', markersize=4, label='Data')

# Perform linear regression
X = sm.add_constant(data['DATE'])  # Add a constant term for the intercept
model = sm.OLS(data['personal_consumption'], X)
results = model.fit()

# Get the fitted values and confidence intervals
fitted_values = results.predict(X)
ci_low, ci_high = results.conf_int().T  # Transpose to get lower and upper bounds

# Plot the fitted line and confidence intervals
plt.plot(data['DATE'], fitted_values, color='blue', linestyle='-', label='Trend Line')
#plt.fill_between(data['DATE'], ci_low, ci_high, color='red', alpha=0.3)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Personal Consumption (fraction of GDP)')
plt.title('Trend of Personal Consumption with Confidence Interval')

# Show legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
